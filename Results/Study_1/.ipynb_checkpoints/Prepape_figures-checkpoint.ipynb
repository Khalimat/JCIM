# Make radar charts

```python
def get_median_for_metric(path, metric_name):
    summary_stats = []
    directories = []
    for directory in Path(working_dir).iterdir():
        if directory.is_dir() and str(directory).split('/')[-1].split('_')[0] in ['ADASYN', 'CNN', 'IHT']:
            directories.append(str(directory).split('/')[-1])
            ds_r = pd.read_csv(directory/ 'DeepSCAMs.csv')[metric_name].mean()
            AL = pd.read_csv(directory/ 'TF_ML_AL.csv')[metric_name].mean()
            non_AL = pd.read_csv(directory / 'TF_ML_non_AL.csv')[metric_name].mean()
            summary_stats.append([ds_r, AL, non_AL])
    result = pd.DataFrame(summary_stats, columns=['DeepSCAMs', 'AL TF MLP', 'non AL TF MLP'], 
                index=directories)
    return result
```

```python
def make_radar_chart(df, name, parent_dir):
    """
    Make a radar chart with performance of AL and non-AL models
    """
    AL = df['AL TF MLP']
    non_AL = df['non AL TF MLP']
    DeepSCAM = df['DeepSCAMs']
    r_theta = df.index
    # Make adaptive max score
    max_perf = 1
    max_perf = max(max(AL), max(non_AL))
    radar = go.Figure()
    radar.add_trace(go.Scatterpolar(
        r=AL,
        theta=r_theta,
        fill='toself',
        name='AL TF MLP'
    ))
    radar.add_trace(go.Scatterpolar(
        r=non_AL,
        theta=r_theta,
        fill='toself',
        name='non-AL TF MLP'
    ))
    radar.add_trace(go.Scatterpolar(
        r=DeepSCAM,
        theta=r_theta,
        fill='toself',
        name='DeepSCAMs'
    ))
#     radar.update_layout(
#         showlegend=True
#     )
    return radar
#     radar_plot_path_svg =  parent_dir / '{}.svg'.format(name)
#     radar_plot_path_png = parent_dir / '{}.png'.format(name)
#     radar.write_image(str(radar_plot_path_svg))
#     radar.write_image(str(radar_plot_path_png))
```
